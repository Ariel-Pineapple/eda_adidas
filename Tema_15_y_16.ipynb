## El análisis de datos

El análisis de datos es el proceso de inspeccionar, limpiar, transformar y modelar datos con el objetivo de descubrir información útil, llegar a conclusiones y apoyar la toma de decisiones.



---


**¿Por Qué es Importante?**

El análisis de datos es crucial porque transforma los datos crudos en conocimiento práctico. Su importancia radica en que permite:

* **Tomar decisiones informadas:** En lugar de basarte en la intuición, usas evidencia sólida para decidir. Por ejemplo, una tienda puede analizar sus ventas para saber qué productos reabastecer y cuáles poner en oferta.

* **Identificar problemas y oportunidades:** Ayuda a detectar cuellos de botella en un proceso, como una página web donde los clientes abandonan la compra, o a encontrar nuevos mercados potenciales.

* **Entender el comportamiento del cliente:** Permite saber qué prefieren tus clientes, cuándo compran y cómo interactúan con tu servicio, lo que ayuda a personalizar la experiencia.

* **Medir el rendimiento:** Facilita el seguimiento de objetivos y métricas clave (KPIs) para evaluar el éxito de una estrategia.

Sin un buen análisis, las empresas y organizaciones operan a ciegas, perdiendo eficiencia, dinero y oportunidades.

funcionalidades de los dos módulos más importantes para manejar datos que

# Numpy

NumPy (Numerical Python) es una biblioteca fundamental de Python para computación científica y análisis numérico. Es especialmente útil para trabajar con arreglos multidimensionales (arrays) y ofrece una amplia colección de funciones matemáticas para operar sobre ellos.

**Características principales de NumPy:**

* Arreglos multidimensionales (ndarray)

* Permite almacenar y manipular datos de manera eficiente en forma de vectores (1D), matrices (2D) o tensores (nD).

* Es mucho más rápido que las listas de Python para operaciones numéricas.

**Operaciones vectorizadas**

* Evita el uso de bucles for al aplicar operaciones directamente sobre los arreglos.


**Funciones matemáticas optimizadas**

Incluye operaciones como:

* Álgebra lineal (np.dot, np.linalg.inv).

* Estadística (np.mean, np.std).

* Trigonometría (np.sin, np.cos).

* Generación de números aleatorios (np.random).

**Broadcasting**

* Permite operar arreglos de diferentes tamaños automáticamente.

* Integración con otras bibliotecas

**Es la base de herramientas como Pandas, SciPy, Matplotlib y scikit-learn.**

```python
from random import randint
class Dado:
    """Clase que representa a un dado."""
    def __init__(self, num_lados=6):
        """Se consideran a un dado con 6 lados."""
        self.num_lados = num_lados
    def lanzar(self):
        """"Devuelve un valor aleatorio entre 1 y el número de lados."""
        return randint(1, self.num_lados)
```

```python
#Se crea un dado de seis lados
dado = Dado()
#Se realizan algunos lanzamientos y se guardan en una lista.
resultados = []
for num_lanzamiento in range(100):
    resultado = dado.lanzar()
    resultados.append(resultado)
print(resultados)
```

```python
from plotly.graph_objs import Bar, Layout
from plotly import offline
#Análisis de los resultados.
frecuencias = []
for valor in range(1, dado.num_lados+1):
    frecuencia = resultados.count(valor)
    frecuencias.append(frecuencia)

#Visualización de los resultados
x_valores = list(range(1, dado.num_lados+1))
datos = [Bar(x=x_valores, y=frecuencias)]

x_eje_config = {'title': 'Resultado'}
y_eje_config = {'title': 'Frecuencia del resultado'}
my_layout = Layout(title='Resultados de lanzar el dado 1000 veces',

xaxis=x_eje_config, yaxis=y_eje_config)

offline.plot({'data': datos, 'layout': my_layout}, filename='Dado.html')
```

### Funciones Artméticas




```python
# substract(): resta los elementos de dos arrays elemento por elemento.

array1 = np.array([4, 90, 15])
array2 = np.array([6, 38, 10])
resultado = np.subtract(array1, array2)
print(resultado)
#
```

```python
# add(): suma los elementos de dos arrays elemento por elemento.

array1 = np.array([4, 90, 15])
array2 = np.array([6, 38, 10])
resultado = np.add(array1, array2)
print(resultado)
```

```python
# Multiply(): multiplica los elementos de dos arrays elemento por elemento.

array1 = np.array([4, 90, 15])
array2 = np.array([6, 38, 10])
resultado = np.multiply(array1, array2)
print(resultado)
```

```python
# divide(): divide los elementos de dos arrays elemento por elemento.

array1 = np.array([4, 90, 15])
array2 = np.array([6, 38, 10])
resultado = np.divide(array1, array2)
print(resultado)
```

```python
# sqrt()

array7 = np.array([4, 90, 15])
resultado = np.sqrt(array7)
print(resultado)
```

```python
# square()

array8 = np.array([4, 90, 15])
resultado = np.square(array8)
print(resultado)
```

```python
# power()

array9 = np.array([4, 90, 15])
array10 = np.array([2, 3, 2])
resultado = np.power(array9, array10)
print(resultado)
```

```python
# gcd(): Devolverl el máximo común divisor

array11 = np.array([4, 90, 15])
array12 = np.array([6, 38, 10])
resultado = np.gcd(array11, array12)
print(resultado)
```

```python
# lcm() devuelve el mínicmo común múltiplo de los elementos de dos arrays

array13 = np.array([4, 90, 15])
array14 = np.array([6, 38, 10])
resultado = np.lcm(array13, array14)
print(resultado)
```

```python
import numpy as np

# Ventas mensuales (ene-dic) para 3 productos: A, B, C
ventas = np.array([
    [120, 95, 80],   # Enero
    [130, 110, 85],  # Febrero
    [115, 90, 78],   # Marzo
    [140, 105, 92],  # Abril
    [160, 150, 110], # Mayo
    [180, 170, 130], # Junio
    [200, 190, 150], # Julio
    [190, 180, 145], # Agosto
    [175, 160, 135], # Septiembre
    [155, 140, 120], # Octubre
    [130, 120, 100], # Noviembre
    [110, 100, 90]   # Diciembre
])
```

```python
# 1. Promedio anual por producto
promedio_anual = np.mean(ventas, axis=0)
print("Promedio anual (A, B, C):", promedio_anual)

# 2. Mes con mayores ventas por producto
mes_max_A = np.argmax(ventas[:, 0])  # Índice del mes (0: enero)
mes_max_B = np.argmax(ventas[:, 1])
mes_max_C = np.argmax(ventas[:, 2])
print(f"Mayores ventas: A (mes {mes_max_A+1}), B (mes {mes_max_B+1}), C (mes {mes_max_C+1})")

# 3. Filtrar meses donde A > 150k
meses_altos_A = ventas[ventas[:, 0] > 150]
print("Meses con A > $150k:\n", meses_altos_A)

# 4. Desviación estándar por producto
std_ventas = np.std(ventas, axis=0)
print("Desviación estándar (A, B, C):", std_ventas)

# 5. Normalizar ventas de B (escalar entre 0 y 1)
ventas_B_normalizadas = (ventas[:, 1] - np.min(ventas[:, 1])) / (np.max(ventas[:, 1]) - np.min(ventas[:, 1]))
print("Ventas B normalizadas:", ventas_B_normalizadas)
```
